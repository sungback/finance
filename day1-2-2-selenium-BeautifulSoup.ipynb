```python
import time

import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
```

```python
url = "https://www.dhlottery.co.kr/lt645/result"
```

```python
# Chrome 브라우저 설정
options = Options()
options.add_argument("--window-size=1280,900")

# 브라우저 창을 띄우지 않고 실행하려면 아래 줄의 #을 지우세요.
# options.add_argument("--headless=new")
```

```python
# Selenium으로 사이트를 열고, 자바스크립트가 만든 HTML을 가져온다.
driver = webdriver.Chrome(options=options)

try:
    driver.get(url)

    # 당첨번호 공이 화면에 생길 때까지 최대 15초 기다린다.
    wait = WebDriverWait(driver, 15)
    wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, ".swiper-slide-active .result-ball")))

    # 조금 더 기다리면 화면 요소가 안정적으로 잡힌다.
    time.sleep(1)

    html = driver.page_source
finally:
    driver.quit()

html[:500]
```

```python
# BeautifulSoup로 HTML 파싱 준비
soup = BeautifulSoup(html, "html.parser")
```

```python
slide = soup.select_one('.swiper-slide-active')
slide
```

```python
draw_no = slide.select_one(".ltEpsd").get_text() # <span class="color-g ltEpsd">1222</span>
draw_no
```

```python
draw_date = slide.select_one(".result-date").get_text() # <div class="result-date">2026.04.25 추첨</div>
draw_date
```

```python
all_divs = slide.select(".result-ball") # 마지막 7번째는 보너스 번호
all_divs
```

```python
all_balls = []
for i in all_divs:
    all_balls.append( i.get_text() )
all_balls
```

```python
balls = all_balls[:6] # [0:6] ==> [0은 포함 : 6은 안 포함] ==> [0,1,2,3,4,5] 총 6개
balls
```

```python
bonus = all_balls[6]
bonus
```

```python
print( draw_no, "회")
print( draw_date )
print("보너스 번호:", bonus)
```
