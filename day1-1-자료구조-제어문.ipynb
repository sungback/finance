```python
# 미션 1 : 점수(0~100)를 가지고 학점(90점이상A,80점이상B,...)을 출력하시오.
# 분석 4줄을 완성하시오. 요리의 레시피처럼 작업의 순서를 나열하면 됨.
# 1. 점수를 저장할 변수를 정의 : score = 77
# 2. 학점을 저장할 변수를 정의 : grade = ""
# 3. if문으로 점수(score)가 90점 이상이면 학점 변수(grade)에 "A"를 저장한다. ...
# 4. 출력 형식 지정 : 예) "77점은 C학점입니다."
```

```python
score = 77
grade = ""
if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"
print(score, "점은", grade, "학점입니다.")
```

```python
# while 문의 구조 : 초기값, 조건, 증가
# while문을 사용하여 "hello world" 를 5번 출력
i = 1 # 1. 초기값
while i <= 5: # 2. 조건 : 1부터 5까지
    print("hello world", i)
    i = i + 1 # 3. 증가
```

```python
# 미션 2 : while문을 사용해서 1부터 100까지의 합을 출력하시오.
# 분석 5줄 : 작업의 순서(레시피) 나열, 명사, 동사(제어문), 출력 형식
# 1. 1부터 100까지 변하는 변수 : i = 1
# 2. 합을 저장할 변수 : total = 0
# 3. while 문의 구조 : 1. 초기값 i = 1, 2. 조건 while i <= 100, 3. 증가 i = i + 1
# 4. 할 일 : 합계를 누적, total = tatal + i
# 5. 출력 형식 : 예) "1부터 100까지의 합은 XXXX입니다."
```

```python
i = 1
total = 0
while i <= 100:
    total = total + i
    i = i + 1
print("1부터 100까지의 합은", total, "입니다.")
```
